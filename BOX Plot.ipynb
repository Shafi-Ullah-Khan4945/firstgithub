```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
titanic= sns.load_dataset("titanic")

sns.boxplot(x="survived", y="age",showmeans=True, data =titanic, meanprops={"marker":"+","markersize":"10","markeredgecolor":"red"})

plt.xlabel("How many Survived",size=15),
plt.ylabel("How much was the age", size=15),
plt.title("Bar plot of statistics", size=20,weight= "bold"),
```
